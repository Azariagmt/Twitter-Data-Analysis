```python
import pandas as pd
import numpy as np

import re
import string

import spacy

import gensim
from gensim import corpora

# libraries for visualization
import pyLDAvis
# import pyLDAvis.gensim
import pyLDAvis.gensim_models as gensimvis
pyLDAvis.enable_notebook()
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
doc_term_matrix = pd.read_pickle('/content/dtm.pkl')
# dictionary = corpora.Dictionary(doc_term_matirx)
doc_term_matirx.head()
```

```python
data_clean = pd.read_pickle('/content/data_clean.pkl')
# dictionary = corpora.Dictionary(doc_term_matirx)
print(data_clean.shape)
data_clean.head()
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
# nlp = spacy.load('en_core_web_md', disable=['parser', 'ner'])

def lemmatization(texts,allowed_postags=['NOUN', 'ADJ']): 
       output = []
       for sent in texts:
             doc = nlp(sent) 
             output.append([token.lemma_ for token in doc if token.pos_ in allowed_postags ])
       return output
```

```python

text_list=data_clean['clean_text'].tolist()
print(text_list[1])
tokenized_reviews = lemmatization(text_list)
print(tokenized_reviews[1])
```

```python
print(tokenized_reviews[1])
```

```python

dictionary = corpora.Dictionary(tokenized_reviews)
doc_term_matrix = [dictionary.doc2bow(rev) for rev in tokenized_reviews]
```

```python
# Creating the object for LDA model using gensim library
LDA = gensim.models.ldamodel.LdaModel

# Build LDA model
lda_model = LDA(corpus=doc_term_matrix, id2word=dictionary, num_topics=10, random_state=100,
                chunksize=100, passes=50,iterations=100)
```

```python
lda_model.print_topics()
```

```python
# Visualize the topics
#https://github.com/bmabey/pyLDAvis
#https://speakerdeck.com/bmabey/visualizing-topic-models
pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim_models.prepare(lda_model, doc_term_matrix, dictionary)
vis
```
